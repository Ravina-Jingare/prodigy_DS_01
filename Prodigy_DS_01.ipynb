```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data of ages in a population

ages = np.random.randint(0, 100, 1000)  # Generate random ages between 0 and 100 for 1000 people
```

```python
# Create histogram

plt.figure(figsize=(10, 6))
plt.hist(ages, bins=20, color='orange', edgecolor='black')

# Adding title and labels

plt.title('Distribution of Ages in a Population')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show plot

plt.show()
```
